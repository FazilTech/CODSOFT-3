# Importing the Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score
sns.set()
```

# Import DataSet

```python
raw_data = pd.read_csv("IRIS.csv")
print(raw_data.head())
```

```python
raw_data.describe()
```

# Check for NULL values

```python
raw_data.isnull().sum()
```

```python
# no null values
data = raw_data.copy()
data
```

# Visualize the DataSet

```python
sns.pairplot(raw_data, hue='species')
plt.show()
```

# Spliting of Datas

```python
inputs = data.drop('species', axis=1)
target = data['species']
inputs
target
```

```python
x_train, x_test, y_train, y_test = train_test_split(inputs, target, test_size=0.3, random_state = 2)
inputs.shape, x_train.shape, x_test.shape
```

# Training the DataSet

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
```

# Prediction for Training Datas

```python
# training the train set datas
x_train_prediction = knn.predict(x_train)
x_train_prediction
```

```python
# Checking the accuracy of the train set
train_accuracy = accuracy_score(y_train, x_train_prediction)
train_accuracy
```

# Prediction for Test Dataset

```python
# Training the Test dataSet
x_test_prediction = knn.predict(x_test)
x_test_prediction
```

```python
# Checking the accuracy score
test_accuracy = accuracy_score(y_test, x_test_prediction)
test_accuracy
```

# Classification Report

```python
print(classification_report(y_test, x_test_prediction))
```
